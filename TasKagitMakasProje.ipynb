```python
import random
import tkinter as tk
from tkinter import messagebox

# Oyun sayaçları
oyuncu_galibiyet = 0
bilgisayar_galibiyet = 0
tur = 0
oyun_devam = True

# Seçenekler listesi
secenekler = ["taş", "kağıt", "makas"]

# Kullanıcı seçimi ve bilgisayar seçimi etiketleri
sonuc_label = None

# Yapay zeka davranışı için önceki oyuncu hamlelerini saklayacak liste
oyuncu_hamleleri = []

# Yapay zeka kişilikli mesajlar
duygusal_mesajlar = {
    'kazanma': [
        "Harika! Senin zaferin kesinlikle hak ediliyor!",
        "Tebrikler! Çok iyi oynadın!",
        "Zafere ulaşmak müthiş bir duygu!"
    ],
    'kaybetme': [
        "Üzgünüm, bu sefer olmadı. Ama çabalarını takdir ediyorum.",
        "Her zaman bir sonraki sefere daha iyi olabilirsin!",
        "Yapay zeka seni bu sefer yendi, ama pes etme!"
    ],
    'beraberlik': [
        "Beraberlik! Hem sen hem de ben bu turda eşit güçteydik.",
        "Güzel bir mücadeleydi. Her ikimiz de aynı skoru aldık!",
        "Beraberlik? İkimiz için, şerefe!"
    ]
}

def yapay_zeka_mesaj(durum):
    """Oyuncu ile esprili bir şekilde sohbet eden yapay zeka."""
    if durum == 'kazanma':
        return random.choice(duygusal_mesajlar['kazanma'])
    elif durum == 'kaybetme':
        return random.choice(duygusal_mesajlar['kaybetme'])
    else:
        return random.choice(duygusal_mesajlar['beraberlik'])

def yapay_zeka_secimi():
    """Oyuncunun geçmiş hamlelerine göre yapay zekanın bir seçim yapmasını sağlar."""
    if not oyuncu_hamleleri:
        return random.choice(secenekler)
    else:
        son_hamle = oyuncu_hamleleri[-1]
        if son_hamle == "taş":
            return "kağıt"  # Kağıt taşı yener
        elif son_hamle == "kağıt":
            return "makas"  # Makas kağıdı yener
        else:
            return "taş"  # Taş makası yener

def yeni_oyun_sorusu():
    """Kullanıcıya ve bilgisayara yeni oyun oynayıp oynamayacaklarını sorar."""
    global oyuncu_galibiyet, bilgisayar_galibiyet, tur, oyun_devam, window
    
    oyuncu_galibiyet = 0
    bilgisayar_galibiyet = 0
    tur = 0
    oyun_devam = True

    if sonuc_label:
        sonuc_label.config(text="Yeni oyun başlıyor. Seçiminizi yapın!")
    oyuncu_hamleleri.clear()  # Yapay zeka için geçmiş hamleler sıfırlanıyor

def sonuc_goster(oyuncu_secimi):
    """Bir tur sonucunu hesaplayıp ekrana gösterir."""
    global oyuncu_galibiyet, bilgisayar_galibiyet, tur, oyun_devam

    if not oyun_devam:
        return
    
    # Bilgisayarın seçimi (yapay zeka ile)
    bilgisayar_secimi = yapay_zeka_secimi()

    # Oyuncu ve bilgisayarın seçimlerini ekrana yazdır
    sonuc_text = f"Oyuncu: {oyuncu_secimi}, Bilgisayar: {bilgisayar_secimi}\n"
    
    # Sonucu belirleme
    if oyuncu_secimi == bilgisayar_secimi:
        sonuc_text += "Beraberlik! Yapay zeka da güçlü görünüyor!"
        oyuncu_galibiyet += 1  # Beraberlikte iki taraf da puan alacak
        bilgisayar_galibiyet += 1
        durum = 'beraberlik'
    elif (oyuncu_secimi == "taş" and bilgisayar_secimi == "makas") or \
         (oyuncu_secimi == "makas" and bilgisayar_secimi == "kağıt") or \
         (oyuncu_secimi == "kağıt" and bilgisayar_secimi == "taş"):
        sonuc_text += "Tebrikler! Kazandınız! Yapay zeka yenildi!"
        oyuncu_galibiyet += 1
        durum = 'kazanma'
    else:
        sonuc_text += "Bilgisayar kazandı! Yapay zeka sizi okudu!"
        bilgisayar_galibiyet += 1
        durum = 'kaybetme'

    # Yapay zeka kişilikli mesaj ekle
    sonuc_text += f"\nYapay Zeka: {yapay_zeka_mesaj(durum)}"
    
    # Oyuncunun hamlesini kaydet
    oyuncu_hamleleri.append(oyuncu_secimi)

    tur += 1
    sonuc_text += f"\nSkor: Oyuncu {oyuncu_galibiyet} - Bilgisayar {bilgisayar_galibiyet}"

    # Sonucu ekranda göster
    sonuc_label.config(text=sonuc_text)

    # İlk iki turu kazananı kontrol et
    if tur >= 2:
        kazanan_belirle()

def kazanan_belirle():
    """İki turdan sonra kazananı belirler ve oyunu sonlandırır."""
    global oyun_devam, window
    
    if oyuncu_galibiyet > bilgisayar_galibiyet:
        messagebox.showinfo("Oyun Bitti", "Tebrikler! Oyunu kazandınız. Yapay zekayı yendiniz!")
    elif bilgisayar_galibiyet > oyuncu_galibiyet:
        messagebox.showinfo("Oyun Bitti", "Bilgisayar oyunu kazandı! Yapay zeka bir adım önde.")
    else:
        messagebox.showinfo("Oyun Bitti", "Oyun berabere bitti! Yapay zeka size eşlik ediyor!")

    oyun_devam = False
    tekrar = messagebox.askyesno("Tekrar Oynamak İster Misiniz?", "Yeni bir oyun oynamak ister misiniz?")
    if tekrar:
        yeni_oyun_sorusu()
    else:
        messagebox.showinfo("Teşekkürler", "Oynadığınız için teşekkürler!")
        window.quit()  # Tkinter penceresini kapatır ve uygulamayı sonlandırır.

def baslangic_ekrani():
    """Başlangıç ekranı: Oyun kuralları ve Oyna butonu."""
    global window
    
    # Pencereyi oluştur
    window = tk.Tk()
    window.title("Taş, Kağıt, Makas Oyunu")
    window.geometry("600x400")
    window.config(bg="lightgray")
    
    # Oyun kurallarını açıklayan yazı
    kurallar_label = tk.Label(
        window, text="Oyun Kuralları:\n\n- Taş makası yener, kağıt taşı yener, makas kağıdı yener.\n"
                     "- İlk iki turu kazanan galip gelir.\n"
                     "- Oyunu kapatmak için pencereyi kapatabilirsiniz.",
        font=("Arial", 16), bg="lightgray"
    )
    kurallar_label.pack(pady=50)
    
    # "Oyna" butonu
    oyna_button = tk.Button(
        window, text="Oyna!", command=tas_kagit_makas_Kagan_Agin_Aksoy,
        font=("Arial", 20), width=10, height=2, bg="white", fg="black"
    )
    oyna_button.pack()

    window.mainloop()

def tas_kagit_makas_Kagan_Agin_Aksoy():
    """Oyunu başlatan ana fonksiyon."""
    global window, sonuc_label
    
    # Eğer pencere mevcutsa, kapat
    if window:
        window.destroy()
    
    # Yeni oyun ekranı
    window = tk.Tk()
    window.title("Taş, Kağıt, Makas")
    window.geometry("600x400")
    window.config(bg="lightgray")
    
    # Sonuç etiketini oluştur
    sonuc_label = tk.Label(
        window, text="Seçiminizi yapın: Taş, Kağıt veya Makas",
        font=("Arial", 16), bg="lightgray"
    )
    sonuc_label.pack(pady=50)
    
    # Taş butonu
    tas_button = tk.Button(
        window, text="Taş", command=lambda: sonuc_goster("taş"),
        font=("Arial", 20), width=10, height=2, bg="white", fg="black"
    )
    tas_button.pack(side=tk.LEFT, padx=10)

    # Kağıt butonu
    kagit_button = tk.Button(
        window, text="Kağıt", command=lambda: sonuc_goster("kağıt"),
        font=("Arial", 20), width=10, height=2, bg="white", fg="black"
    )
    kagit_button.pack(side=tk.LEFT, padx=10)

    # Makas butonu
    makas_button = tk.Button(
        window, text="Makas", command=lambda: sonuc_goster("makas"),
        font=("Arial", 20), width=10, height=2, bg="white", fg="black"
    )
    makas_button.pack(side=tk.LEFT, padx=10)
    
    window.mainloop()

def terminal_versiyonu():
    """Terminal üzerinden oyunun çalışmasını sağlar."""
    global oyuncu_galibiyet, bilgisayar_galibiyet, tur, oyun_devam

    def oyun_bitti():
        if oyuncu_galibiyet > bilgisayar_galibiyet:
            print("Tebrikler! Oyunu kazandınız. Yapay zekayı yendiniz!")
        elif bilgisayar_galibiyet > oyuncu_galibiyet:
            print("Bilgisayar oyunu kazandı! Yapay zeka bir adım önde.")
        else:
            print("Oyun berabere bitti! Yapay zeka size eşlik ediyor!")

    def secim_yap():
        global tur, oyuncu_galibiyet, bilgisayar_galibiyet
        while tur < 2:
            oyuncu_secimi = input("Seçiminizi yapın (taş, kağıt, makas): ").strip().lower()
            if oyuncu_secimi not in secenekler:
                print("Geçersiz seçim. Lütfen 'taş', 'kağıt' veya 'makas' girin.")
                continue

            bilgisayar_secimi = yapay_zeka_secimi()
            print(f"Oyuncu: {oyuncu_secimi}, Bilgisayar: {bilgisayar_secimi}")

            if oyuncu_secimi == bilgisayar_secimi:
                print("Beraberlik! Yapay zeka da güçlü görünüyor!")
                oyuncu_galibiyet += 1
                bilgisayar_galibiyet += 1
            elif (oyuncu_secimi == "taş" and bilgisayar_secimi == "makas") or \
                 (oyuncu_secimi == "makas" and bilgisayar_secimi == "kağıt") or \
                 (oyuncu_secimi == "kağıt" and bilgisayar_secimi == "taş"):
                print("Tebrikler! Kazandınız! Yapay zeka yenildi!")
                oyuncu_galibiyet += 1
            else:
                print("Bilgisayar kazandı! Yapay zeka sizi okudu!")
                bilgisayar_galibiyet += 1

            tur += 1
            print(f"Skor: Oyuncu {oyuncu_galibiyet} - Bilgisayar {bilgisayar_galibiyet}")

        oyun_bitti()

    while True:
        secim_yap()
        tekrar = input("Yeni bir oyun oynamak ister misiniz? (e/h): ").strip().lower()
        if tekrar != 'e':
            print("Oynadığınız için teşekkürler!")
            break
        oyuncu_galibiyet = 0
        bilgisayar_galibiyet = 0
        tur = 0
        oyun_devam = True

if __name__ == "__main__":
    mod = input("GUI mi yoksa Terminal mi? (g/t): ").strip().lower()
    if mod == 'g':
        baslangic_ekrani()
    elif mod == 't':
        terminal_versiyonu()
    else:
        print("Geçersiz seçenek. 'g' (GUI) veya 't' (Terminal) girin.")
```
